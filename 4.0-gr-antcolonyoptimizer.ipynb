```python
import tsplib95
import numpy as np
import matplotlib.pyplot as plt
```

```python
#importing the data sets with the cities and their coordinates / type of distance

problem1 = tsplib95.load('ch130.tsp')
problem2 = tsplib95.load('tsp225.tsp')
```

```python
#function that calculates the euclidian distance between two nodes (given their coordinates)

def distance(start, end):
    x_distance = abs(start[0] - end[0])
    y_distance = abs(start[1] - end[1])

    import math
    return math.sqrt(pow(x_distance, 2) + pow(y_distance, 2))
```

```python
#function that returns the matrix with the distances between the cities of the second data set - 
#utilizing the list of coordinates given

def mpl(M,n):
    i=0
    while i<n:
        aux=[]
        for j in range(0,n):
            if i==j:
                aux.append(99999)
            else:
                aux.append(distance(problem2.node_coords[i+1],problem2.node_coords[j+1]))
        
        M[i]=aux
        i=i+1
    return M


#matrix with the distances
F=[[0 for i in range(0,225)]]*225
F1=mpl(F,225)
F2=np.array(F1)

```

```python
#function that returns the matrix with the distances between the cities of the second data set - 
#utilizing the list of coordinates given


def mpl1(M,n):
    i=0
    while i<n:
        aux=[]
        for j in range(0,n):
            if i==j:
                aux.append(99999)
            else:
                aux.append(distance(problem1.node_coords[i+1],problem1.node_coords[j+1]))
        
        M[i]=aux
        i=i+1
    return M

#matrix with the distances

B=[[0 for i in range(0,130)]]*130
B1=mpl1(B,130)
B2=np.array(B1)
```

```python
#importing the ant colony function thats actually going to be used
#o algoritmo foi retirado deste site https://github.com/johnberroa/Ant-Colony-Optimization

from AntColonyOptimizer import AntColonyOptimizer
```

```python
#plotting the solutions of ch130 obtained for each iterate with normal values of the parameters~nad its relative errors

optimizer = AntColonyOptimizer(ants=20, evaporation_rate=.005, intensification=1, alpha=1, beta=7,
                               beta_evaporation_rate=0, choose_best=.1)
 
best = optimizer.fit(B2, 500)
t=optimizer.best_series
x=[i for i in range(0,500)]
plt.plot(x, t, color='r', label='best=best')
plt.xlabel("Iterations")
plt.ylabel("Travelled distance")
plt.title("ch130")
plt.rcParams['figure.figsize'] = [5,5]
plt.show()

t2=[]
for i in range(0,500):
    t2.append(abs(6110-t[i])/6110)
    
plt.plot(x,t2,color='b')
plt.xlabel("Iterations")
plt.ylabel("Relative errors")
plt.title("ch130")
plt.rcParams['figure.figsize'] = [5,5]
plt.show()
```
